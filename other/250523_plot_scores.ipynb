```python
import matplotlib.pyplot as plt

# Data
sent = 'a cream colored and black cat stands on a refrigerator shelf as a hand holds some fruit'
words = ['hand', 'cream', 'black', 'on', 'some', 'cat', 'fruit', 'refrigerator', 'shelf', 'colored']
concreteness = [4.72, 4.83, 3.76, 3.25, 2.48, 4.86, 4.81, 5.0, 4.96, 4.0]
# Sort words by their concreteness scores in descending order
sorted_words = sorted(zip(words, concreteness), key=lambda x: x[1], reverse=True)
print(sorted_words)

words_list = []
scores_list = []
for word in sent.split(' '):
    words_list.append(word)
    if word in words:
        scores_list.append(concreteness[words.index(word)])
    else:
        scores_list.append(0)

# Plotting
plt.figure(figsize=(10, 2))
plt.bar(words_list, scores_list, color='#0c74c4')
# plt.xlabel('Words')
plt.ylabel('Concreteness Score')
# plt.title('Concreteness Scores of Words')
plt.xticks(rotation=45)
plt.tight_layout()
plt.xticks(rotation=0)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().tick_params(left=False, bottom=False)
plt.gca().yaxis.set_visible(False)
plt.tight_layout()
plt.show()
```
